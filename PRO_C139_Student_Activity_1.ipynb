```python
# creating dataframe
import pandas as pd
credits_df = pd.read_csv('/content/kaggle-movie-data/tmdb_5000_credits.csv')
movies_df = pd.read_csv('/content/kaggle-movie-data/tmdb_5000_movies.csv')
```

```python
# printing first 5 rows
credits_df.head()
```

```python
# printing first 5 rows
movies_df.head()
```

```python
# renaming a column
credits_df.rename(columns = {'movie_id' : 'id'} , inplace = True)
credits_df.head()
```

```python
# merging datasets
result_df = pd.merge(movies_df , credits_df , on = 'id')
```

```python
# printing shape of dataset
result_df.shape
```

```python
# print the dataframe info
result_df.info()
```

```python
# clean the dataset : remove the columns
result_df.drop(['homepage', 'title_x', 'title_y', 'production_companies'] , axis = 1 , inplace = True)
```

```python
# print the dataframe info again
result_df.info()
```

```python
# dropping cols with na
result_df.dropna(inplace = True)
result_df.info()
```

```python
# Calculating weighted average : the value of R
R = result_df['vote_average']
```

```python
# Calculating weighted average : the value of v
v = result_df['vote_count']
```

```python
# Calculating weighted average : the value of C
C = result_df['vote_average'].mean()
print(C)
```

```python
# Calculating weighted average : the value of m
m = result_df['vote_count'].quantile(0.9)
print(m)
```

```python
# adding weighted_rating column
result_df['weighted_rating'] = (R*v+C*m)/(v+m)
```

```python
# print the first 5 rows
result_df[['original_title','weighted_rating']].head()
```

```python
# sort the values : descending order
result_df.sort_values('weighted_rating',ascending = False, inplace = True)
```

```python
# print the first 10 rows
result_df[['original_title','weighted_rating']].head(10)
```

```python
# plot it
import plotly.express as px
bar_plot = px.bar(result_df.head(10).sort_values('weighted_rating'),x = 'weighted_rating', y = 'original_title', orientation='h')
bar_plot.show()
```
